```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/harish/Desktop/LETS UPGRADE/general_data.csv")
data
```

```python
data["NumCompaniesWorked"]=data["NumCompaniesWorked"].fillna(method="pad")
```

```python
data["TotalWorkingYears"]=data["TotalWorkingYears"].fillna(method="pad")
```

```python
data.isnull().sum().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data["Attrition"]=le.fit_transform(data["Attrition"])
```

```python
y=data["Attrition"]
```

```python
X=data[["Age","DistanceFromHome","Education","PercentSalaryHike","YearsAtCompany"]]
```

```python
import statsmodels.api as sm
```

```python
X1=sm.add_constant(X)
```

```python
logistic=sm.Logit(y,X1)
```

```python
result=logistic.fit()
```
